```python
import pandas as pd
import statsmodels.api as sm
```

```python
dta=[10930,10318,10595,10972,7706,6756,9092,10551,9722,10913,11151,8186,6422,6337,11649,11652,10310,12043,7937,6476,9662,9570,9981,9331]
dta=pd.Series(dta)
dta.index=pd.Index(sm.tsa.datetools.dates_from_range('1996','2019'))
print(dta)
```

```python
import matplotlib.pyplot as plt
```

```python
# 对数据绘图，观测是否为平稳时间序列
dta.plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(dta)
```

```python
# 1阶差分，数据平稳化
diff1 = dta.diff(1).dropna()
```

```python
diff1.columns = [u'销量差分']
```

```python
diff1.plot()
```

```python
print(dta,diff1)
```
